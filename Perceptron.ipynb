```python
def train_single_sample(a ,y, labels, n_iterations, lr, variable_lr):
  acc_history = []
  model_list = []
  k = 0
  i = 1
  n = len(y)

  if variable_lr:
    learn_rate = Scheduler(lr, k)
    learn_rate.lr = lr
    learn_rate.factor = 0
  while i < n_iterations:

    if variable_lr:
      if k != 0:
        learn_rate.factor = k
        lr = learn_rate.get_next_lr()

    k = np.random.randint(0, len(y))
    in_prod = float(a.T @ y[k])
    pred = 0

    if in_prod < 0:
      pred += 1

    delta = pred - labels[k]
    err = delta * lr * y[k]
    a = np.add(a, err.reshape((3, 1)))
    model_list.append(a)

    if i % n == 0:
      prediction = a.T @ y.T
      predicted_labels = [0 if pred >= 0 else 1 for pred in prediction[0]]
      accuracy = sum(predicted_labels - labels == 0) / len(labels)
      acc_history.append(accuracy)
      best_choice = np.argmax(acc_history)

    i += 1

  return model_list[best_choice], acc_history
```

```python
class Scheduler:
  def __init__(self, lr, factor):
    self.lr = lr
    self.factor = factor

  def get_next_lr(self):
    self.lr = self.lr / m.sqrt(self.factor)
    return self.lr

learn = Scheduler(10000,  10)
learn_array = [learn.get_next_lr() for i in range(100)]

fig, axis = plt.subplots(1, 2, figsize = (10, 5))
axis[0].plot(learn_array)

axis[1].plot(learn_array)
axis[1].set_yscale("log")
plt.show()
```

```python
def train_batch(a ,y, labels, n_iterations, lr, variable_lr, theta, batch_size):
  acc_history = []
  error_history = []
  model_list = []

  k = 0
  i = 1
  n = len(y)
  sum = 0

  if variable_lr:
    learn_rate = Scheduler(lr, k)
    learn_rate.lr = lr
    learn_rate.factor = 0

  abs_err = theta + 1
  while abs_err > theta:

    if variable_lr:
      if k != 0:
        learn_rate.factor = k
        lr = learn_rate.get_next_lr()

    k_list = np.array([np.random.randint(0, len(y)) for i in range((batch_size))])
    sum_list = np.array([y[k] for k in k_list])
    in_prod = np.array([float(a.T @ y[k]) for k in k_list])
    pred = np.array([0 if x >= 0 else 1 for x in in_prod])

    labels_k = np.array([labels[k] for k in k_list])
    delta = np.add(pred, - labels_k)
    err = lr * delta @ sum_list
    abs_list = [x ** 2 for x in err]
    abs_err = lr * m.sqrt(np.sum(abs_list))
    a = np.add(a, err.reshape((3,1)))
    model_list.append(a)

    if i * n % batch_size == 0:
      prediction = a.T @ y.T
      predicted_labels = [0 if pred >= 0 else 1 for pred in prediction[0]]
      accuracy = np.sum(predicted_labels - labels == 0) / len(labels)
      acc_history.append(accuracy)
      error_history.append(1 - accuracy)
      best_choice = np.argmax(acc_history)

    i += 1

  return model_list[best_choice], acc_history, error_history
```
